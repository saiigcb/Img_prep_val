```python
import cv2
import os
import xml.etree.ElementTree as ET

def show_img(img, name):
    '''
    Used for valuating the image for bounding box
    '''
    imS = cv2.resize(img, (700, 540))
    cv2.imshow(name, imS)
    choice = cv2.waitKey(0)
    if  choice == ord('a'):
        cv2.imwrite('./accept/'+name, imS)
        cv2.destroyAllWindows()

    elif choice == ord('r'):
        cv2.imwrite('./reject/'+name, imS)
        cv2.destroyAllWindows()

def read_content(xml_file: str):
    '''
    read an XML file and parse it collect the necessary fields
    Here BBOX coords and the class of the label are collected
    '''
    tree = ET.parse(xml_file)
    root = tree.getroot()
    list_with_all_boxes = []
    names = []
    combo = {}
    
    for boxes in root.iter('object'):
    
        ymin, xmin, ymax, xmax = None, None, None, None 

        ymin = int(float(boxes.find("bndbox/ymin").text))
        xmin = int(float(boxes.find("bndbox/xmin").text))
        ymax = int(float(boxes.find("bndbox/ymax").text))
        xmax = int(float(boxes.find("bndbox/xmax").text))
        label = boxes.find("name").text
        names.append(label)
        list_with_single_boxes = [xmin, ymin, xmax, ymax]
        list_with_all_boxes.append(list_with_single_boxes)
        combo[label] = list_with_single_boxes
    
    return combo

xml_path = '/home/intellect/projects/Aug_dataset/XML/'
img_path = '/home/intellect/projects/Aug_dataset/images/'
count = 0
for files in os.listdir(xml_path):
    if files.endswith('.xml'):
        with open(os.path.join(xml_path,files), 'r') as f:
            data = f.read()
        xml_file = read_content(os.path.join(xml_path,files))
        file_name = files.split('.')[0]
        img_name = file_name+'.jpg'
        if os.path.exists(os.path.join(img_path,img_name)):
            img_name = file_name+'.jpg'
        if not os.path.exists(os.path.join(img_path,img_name)):
            img_name = file_name+'.PNG'
        
        img = cv2.imread(os.path.join(img_path, img_name))
        
        try:
            colors = [(130,100,180), (250, 150, 120), (50, 255, 255)]
            for i in range(0, len(colors)):
                x1, y1, x2, y2 = list(xml_file.values())[i]
                cv2.rectangle(img, (x1, y1), (x2, y2), colors[i], 2)
                
            
            cv2.imwrite('/home/intellect/projects/Aug_dataset/combo/'+img_name, img)
#             show_img(img, img_name)
            print('accept')
        except:
            path = '/home/intellect/projects/Aug_dataset/'
            print('XML file check: ', files)
            font = cv2.FONT_HERSHEY_SIMPLEX
            labels = list(xml_file.keys())
            colors = [(130,100,180), (250, 150, 120)]
            for i in range(0, len(colors)):
                x1, y1, x2, y2 = list(xml_file.values())[i]
                org = (x1-10, y1-10)
                cv2.rectangle(img, (x1, y1), (x2, y2), colors[i], 2)
                cv2.putText(img, labels[i],org,font,0.6, colors[i], 2, cv2.LINE_AA)

            cv2.imwrite(os.path.join(path,img_name), img)
            count += 1
print(count)
```
